```python
### def calculate_tax():
    print("--- Staff Tax Revenue Assessment System ---")

    try:
        
        experience = int(input("Enter years of experience: "))
        age = int(input("Enter age of the staff: "))

     
        if experience > 25 and age >= 55:
            tax_revenue = 5600000
            print(f"\nResult: Annual Tax Revenue is NGN {tax_revenue:,}")

       
        elif experience > 20 and age >= 45:
            tax_revenue = 4480000
            print(f"\nResult: Annual Tax Revenue is NGN {tax_revenue:,}")

        
        elif experience > 10 and age >= 35:
            tax_revenue = 1500000
            print(f"\nResult: Annual Tax Revenue is NGN {tax_revenue:,}")

        
        else:
            tax_revenue = 550000
            print(f"\nResult: Annual Tax Revenue is NGN {tax_revenue:,}")

    except ValueError:
        print("Error: Please enter valid numeric values for age and experience.")

if __name__ == "__main__":
    calculate_tax()
```
