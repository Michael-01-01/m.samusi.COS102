```python
## import numpy as np

def solver():
    print("--- Polynomial Root Finder ---")
    print("Select the degree of the equation:")
    print("1. Quadratic (Degree 2)")
    print("2. Cubic     (Degree 3)")
    print("3. Quartic   (Degree 4)")

    choice = input("\nEnter choice (1, 2, or 3): ")

    if choice == '1':
        print("\nFormat: ax² + bx + c = 0")
        a = float(input("Enter a: "))
        b = float(input("Enter b: "))
        c = float(input("Enter c: "))
        coeffs = [a, b, c]

    elif choice == '2':
        print("\nFormat: ax³ + bx² + cx + d = 0")
        a = float(input("Enter a: "))
        b = float(input("Enter b: "))
        c = float(input("Enter c: "))
        d = float(input("Enter d: "))
        coeffs = [a, b, c, d]

    elif choice == '3':
        print("\nFormat: ax⁴ + bx³ + cx² + dx + e = 0")
        a = float(input("Enter a: "))
        b = float(input("Enter b: "))
        c = float(input("Enter c: "))
        d = float(input("Enter d: "))
        e = float(input("Enter e: "))
        coeffs = [a, b, c, d, e]

    else:
        print("Invalid choice. Please select 1, 2, or 3.")
        return

   
    roots = np.roots(coeffs)

    print("\n--- Results ---")
    for i, r in enumerate(roots, 1):
        
        print(f"Root {i}: {np.round(r, 4)}")

if __name__ == "__main__":
    solver()
```
