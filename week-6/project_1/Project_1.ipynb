```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee():
    def __init__(self):
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        
        self.tasks = ["Loading", "Transporting", "Reveiwing Orders", "Customer Service", "Delivering Items"]

        self.window = tk.Tk()
        self.window.title("Logistics System")
        self.window.geometry("300x250")

        self.label = tk.Label(self.window, text="Employee Name:")
        self.label.pack(pady=10)

        self.name_entry = tk.Entry(self.window)
        self.name_entry.pack(pady=5)

        self.btn = tk.Button(self.window, text="Verify", command=self.check_employee)
        self.btn.pack(pady=20)

        self.window.mainloop()

    def check_employee(self):
        name = self.name_entry.get()
        if name in self.employees:
            self.take_attendance(name)
        else:
            self.refuse_access()

    def take_attendance(self, name):
        self.assign_task(name)

    def assign_task(self, name):
        index = random.randint(0, 4)
        selected_task = self.tasks[index]
        messagebox.showinfo("Success", f"Attendance taken.\nName: {name}\nTask: {selected_task}")

    def refuse_access(self):
        messagebox.showerror("Error", "Access Denied: Name not found.")

if __name__ == "__main__":
    Employee()
```
