```python
import tkinter as tk
from tkinter import messagebox

class DeliveryService:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Delivery Price Calculator")
        self.root.geometry("300x300")

        self.loc_label = tk.Label(self.root, text="Select Location:")
        self.loc_label.pack(pady=5)

        self.location_var = tk.StringVar(self.root)
        self.location_var.set("PAU") 
        self.location_menu = tk.OptionMenu(self.root, self.location_var, "PAU", "Epe")
        self.location_menu.pack(pady=5)

        self.weight_label = tk.Label(self.root, text="Enter Weight (kg):")
        self.weight_label.pack(pady=5)

        self.weight_entry = tk.Entry(self.root)
        self.weight_entry.pack(pady=5)

        self.calc_btn = tk.Button(self.root, text="Calculate Cost", command=self.calculate_price)
        self.calc_btn.pack(pady=20)

        self.root.mainloop()

    def calculate_price(self):
        try:
            location = self.location_var.get()
            weight = float(self.weight_entry.get())
            price = 0

            if location == "PAU":
                if weight >= 10:
                    price = 2000
                else:
                    price = 1500
            
            elif location == "Epe":
                if weight >= 10:
                    price = 5000
                else:
                    price = 4000

            messagebox.showinfo("Quote", f"Delivery to {location}\nWeight: {weight}kg\nTotal Cost: N{price}")
            
        except ValueError:
            messagebox.showerror("Input Error", "Please enter a valid number for weight")

if __name__ == "__main__":
    DeliveryService()
```
